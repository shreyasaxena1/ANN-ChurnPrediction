```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X = df.drop('Exited',axis=1)
y = df['Exited']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state = 10)
```

```python
X_train.head()
```

```python
lable_encoder = LabelEncoder()
X_train['Gender']=lable_encoder.fit_transform(X_train['Gender'])
X_test['Gender']= lable_encoder.transform(X_test['Gender'])
```

```python
one_hot_encoder = OneHotEncoder()

# Encode
encoded = one_hot_encoder.fit_transform(X_train[['Geography']]).toarray()

# Convert to DataFrame
encoder_df = pd.DataFrame(
    encoded, 
    columns=one_hot_encoder.get_feature_names_out(['Geography']),
    index=X_train.index   # keep row alignment
)

# Concatenate with original
X_train = pd.concat([X_train.drop('Geography', axis=1), encoder_df], axis=1)
```

```python
# Encode
encoded = one_hot_encoder.transform(X_test[['Geography']]).toarray()

# Convert to DataFrame
encoder_df = pd.DataFrame(
    encoded, 
    columns=one_hot_encoder.get_feature_names_out(['Geography']),
    index=X_test.index   # keep row alignment
)

# Concatenate with original
X_test = pd.concat([X_test.drop('Geography', axis=1), encoder_df], axis=1)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
## Save encoder
import pickle

with open('lable_encoder_gender.pkl','wb') as file:
    pickle.dump(lable_encoder,file)

with open('one_hot_encoder_geo.pkl','wb') as file:
    pickle.dump(one_hot_encoder,file)
```

```python
# Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

## ANN Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape = (X_train.shape[1],)), #HL1 connected with input layer
    Dense(32, activation='relu'), #HL2
    Dense(1, activation='sigmoid') #Output layer
])
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
# loss = tf.keras.losses.BinaryCrossentropy()
```

```python
# Compile the model
model.compile(optimizer=opt, loss="binary_crossentropy", metrics=['accuracy'])
```

```python
# Setup tensorboard
log_dir = "log/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
tensorflow_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
## Setup Early stopping - when loss value is not decreasing we don't need to run all the batches and stop training
#minitor loss  - wait for 5 run to take decision - restore best weights
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
## Train the model
history = model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=100,callbacks=[tensorflow_callback,early_stopping_callback])
```

```python
model.save('model.h5')
```
